```python
# prompt: Use person_data.xlsx file and use the three entries to make three cards in the above format. Then download each card separately as pdf file.

import pandas as pd
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
from google.colab import files

# Load data from the Excel file
try:
  df = pd.read_excel('person_data.xlsx')
except FileNotFoundError:
  print("Error: person_data.xlsx file not found.")
  exit()

# Fetch the image from the provided URL
image_url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTMvEbKIGsZCbu1m-rgbPgukFUgHg2SiDfuLw&s"
try:
  response = requests.get(image_url)
  image = Image.open(BytesIO(response.content))
except Exception as e:
  print(f"Error fetching image: {e}")
  image = None


def create_and_download_card(student_name, father_name, roll_number):
  msg = f"""
      PIAIC Student Card

  Student Name: {student_name}
  Father Name: {father_name}
  Roll Number: {roll_number}
  """

  if image:
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].text(0.5, 0.5, msg, ha='center', va='center')
    ax[0].axis('off')
    ax[1].imshow(image)
    ax[1].axis('off')
    plt.show()

    # Save the figure as a PDF
    plt.savefig(f'{student_name}_card.pdf', format='pdf', bbox_inches='tight')

    # Download the PDF file
    files.download(f'{student_name}_card.pdf')


# Iterate through the DataFrame and create cards for each entry
for index, row in df.iterrows():
  create_and_download_card(row['Student Name'], row['Father Name'], row['Roll Number'])
```
